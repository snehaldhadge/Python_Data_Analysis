```python
import numpy as np

countries = np.array(['A','B','C','D','E','F','G','H'])

employment = np.array([45.7,56.2,70.9,32.5,67.9,98.6,21.5,44.8])

#to find max employment and country

def get_max_employment_country(countries,employment):
    i = np.argmax(employment)
    return(countries[i],employment[i])

max_country,max_employment = get_max_employment_country(countries,employment)
print(max_country,max_employment)
```

where i4 corresponds to a 32-bit integer, f4 corresponds to a 32-bit float, and a10 corresponds to a string 10 characters long.

```python

# Creating an array of zeros and defining column types 
recarr = np.zeros((2,), dtype=('i4,f4,a10'))
toadd = [(1,2.,'Hello'),(2,3.,"World")]
recarr[:] = toadd
recarr
```

Same as above using numpy

```python
# Creating an array of zeros and defining column types
recarr = np.zeros((2,), dtype=('i4,f4,a10'))
# Now creating the columns we want to put # in the recarray
col1 = np.arange(2) + 1
col2 = np.arange(2, dtype=np.float32) 
col3 = ['Hello', 'World']
# Here we create a list of tuples that is
# identical to the previous toadd list. 
toadd = zip(col1, col2, col3)
# Assigning values to recarr 
recarr[:] = toadd
recarr
```

```python
import numpy as np
from scipy.optimize import curve_fit
# Creating a function to model and create data 
def func(x, a, b):
    return a * x +b
# Generating clean data
x = np.linspace(0,10,100)
y = func(x, 1, 2)


# Adding noise to the data
yn = y + 0.9 * np.random.normal(size=len(x))
# Executing curve_fit on noisy data 
popt, pcov = curve_fit(func, x, yn)
# popt returns the best fit values for parameters of # the given model (func).
print(popt)
print(yn)
print(y)
```
